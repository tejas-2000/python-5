```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,5)
y=x*10 
y=[i**2 for i in x]
plt.plot(x,y,marker="*",ms="15",mec="w",mfc="k",c="g") #aaaaaa
plt.xlabel('x-axixs')
plt.ylabel('x-squares')
plt.title('EXAMPLE FOR MARKERS')
plt.show()

```

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,5)
y=x*10 
y=[i**2 for i in x]
plt.plot(x,y,ls="dashed",lw=20)
plt.xlabel('x-axixs')
plt.ylabel('x-squares')
plt.title('SIMPLE LINE PLOT')
plt.show()
```
